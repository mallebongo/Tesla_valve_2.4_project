```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
verkleinings_factor = 3
```

```python
cilinder_invoer = [5.092958179,0.568410511,2.233753587,4.244131816,5.092958179,1.107164822,0.402923907,4.897075172,4.390481189,3.441187959,2.767912054,1.818913635,1.414710605]
cilinder_uitvoer = [3.183098862,0.568410511,1.958830069,2.961022197,3.105462304,1.069949197,0.40420303,3.183098862,2.961022197,2.59844805,2.31498099,1.720593979,1.354510154]
cilinder_invoer_fout = np.array(cilinder_invoer)*0.12/verkleinings_factor
cilinder_uitvoer_fout = np.array(cilinder_uitvoer)*0.12/verkleinings_factor
```

```python
kubussen_invoer = [1.697652726,0.499309625,0.832182709,1.079016563,2.448537586,3.350630381,4.107224338,5.092958179]
kubussen_uitvoer = [1.534023548,0.503256737,0.816179195,1.018591636,2.158033127,2.496548127,2.767912054,2.961022197]
kubussen_invoer_fout = np.array(kubussen_invoer)*0.12/verkleinings_factor
kubussen_uitvoer_fout = np.array(kubussen_uitvoer)*0.12/verkleinings_factor
```

```python
prism_invoer = [0.677255077,1.299224025,2.233753587,2.709020308,3.536776513,4.390481189,4.897075172,0.396647833]
prism_uitvoer = [0.670126076,1.236154898,1.929150825,2.31498099,2.767912054,3.105462304,3.264716781,0.395416008]
prism_invoer_fout = np.array(prism_invoer)*0.12/verkleinings_factor
prism_uitvoer_fout = np.array(prism_uitvoer)*0.12/verkleinings_factor
```

```python
comsol_invoer = np.linspace(0,5,10)
comsol_uitvoer = np.linspace(0,5,10)
```

```python
#plt.scatter(prism_invoer,prism_uitvoer,marker='v')
#plt.scatter(kubussen_invoer,kubussen_uitvoer,marker='s')
#plt.scatter(cilinder_invoer,cilinder_uitvoer,marker='o')

plt.errorbar(prism_invoer,prism_uitvoer,prism_invoer_fout,prism_uitvoer_fout,fmt='v',elinewidth=0.5,capsize=2,markersize=5)
plt.errorbar(kubussen_invoer,kubussen_uitvoer,kubussen_invoer_fout,kubussen_uitvoer_fout,fmt='s',elinewidth=0.5,capsize=2,markersize=5)
plt.errorbar(cilinder_invoer,cilinder_uitvoer,cilinder_invoer_fout,cilinder_uitvoer_fout,fmt='o',elinewidth=0.5,capsize=2,markersize=5)

#plt.scatter(comsol_invoer,comsol_uitvoer)
plt.legend(['Prisma','Kubus','Cilinder'])
plt.xlabel('Invoer snelheid [m/s]')
plt.ylabel('Uitvoer snelheid [m/s]')
plt.grid()
plt.title("Invoer snelheid uitgezet tegen uitvoer snelheid van verschillende valves")
plt.text(4,0.4,'Meetfout schaal 1:3',fontsize='small',bbox=dict(boxstyle="round",ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
plt.savefig('3 valves met error bars',dpi=600)
plt.show()
```
